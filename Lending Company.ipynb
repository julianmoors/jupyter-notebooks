### Introduction
_This project loads all loans data and shows whether or not the company is profitable._

```python
import pandas as pd

def format_currency(x):
    return "${:,.2f}".format(x)

df = pd.read_csv('data/lending-company.csv', index_col='LoanID')
pd.set_option('display.max_rows', None)

df['TotalRevenue'] = df['AmtPaid36'] + df['AmtPaid60'] + df['AmtPaid360']
df['TotalCost'] = df['Deposit'] + (df['DailyRate'] * df['TotalDaysYr'])
df['Profit'] = df['TotalRevenue'] - df['TotalCost']
```

```python
df.query('LoanStatus == "Active"')
```

```python
print('Total:', format_currency(df.query('LoanStatus == "Active"')['Profit'].sum()))
```

```python
df.query('LoanStatus == "Finished Payment"')
```

```python
print('Total:', format_currency(df.query('LoanStatus == "Finished Payment"')['Profit'].sum()))
```

```python
df.query('LoanStatus == "Blocked"')
```

```python
print('Total:', format_currency(df.query('LoanStatus == "Blocked"')['Profit'].sum()))
```
